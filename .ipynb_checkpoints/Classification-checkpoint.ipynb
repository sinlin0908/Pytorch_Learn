```python
import torch
import torch.nn.functional as Function

import matplotlib.pyplot as plt

config = {
    'epoch':500,
    'lr': 0.001
}
```

```python
# class Data(object):
#     def __init__(self):
#         n_data = torch.ones(100,2) # (100,2)

#         x0 = torch.normal(2*n_data,1) # means,std的离散正态分布中抽取随机数。 (100,2)
#         y0 = torch.zeros(100) # (100,1)

#         x1 = torch.normal(-2*n_data,1) # (100,2)
#         y1 = torch.ones(100)

#         self.x = torch.cat((x0,x1),0).type(torch.FloatTensor)  # shape (200, 2) FloatTensor = 32-bit floating
#         self.y = torch.cat((y0,y1),).type(torch.LongTensor)  # shape (200,) LongTensor = 64-bit integer

        
# #         self.x ,self.y = Variable(self.x),Variable(self.y)

#     def get_data(self):
#         return self.x ,self.y
n_data = torch.ones(100, 2)
x0 = torch.normal(2*n_data, 1)      # class0 x data (tensor), shape=(100, 2)
y0 = torch.zeros(100)               # class0 y data (tensor), shape=(100, 1)
x1 = torch.normal(-2*n_data, 1)     # class1 x data (tensor), shape=(100, 2)
y1 = torch.ones(100)                # class1 y data (tensor), shape=(100, 1)
x_train = torch.cat((x0, x1), 0).type(torch.FloatTensor)  # shape (200, 2) FloatTensor = 32-bit floating
y_train = torch.cat((y0, y1), ).type(torch.LongTensor)    # shape (200,) LongTensor = 64-bit integer
```

```python
# data = Data()
# x_train,y_train = data.get_data()


plt.scatter(x_train.numpy()[:,0],
            x_train.numpy()[:,1],
            c=y_train.numpy(),
            s=100,
            lw=0,
            cmap='RdYlGn'
           ) 
plt.show()
```

```python
# model 定義

class Model(torch.nn.Module):
    def __init__(self,n_feature=1,n_hidden=1,n_label=1):
        super(Model,self).__init__()
        
        # Define Network
        # 這樣不代表已經建好神經網路
        self.hidden_layer = torch.nn.Linear(n_feature,n_hidden)
        self.output_layer = torch.nn.Linear(n_hidden,n_label)
        
        
    def forward(self,x):
        x = Function.relu(self.hidden_layer(x))
        x = self.output_layer(x)
        
        return x
```

```python
model = Model(n_feature=2, 
              n_hidden=10, 
             n_label=2
             )

print(model)
```

```python
optimizer = torch.optim.Adam(model.parameters(),lr=config['lr'])
loss_func = torch.nn.CrossEntropyLoss() 

model.train()

for step in range(config['epoch']):
    out = Model(x_train)
    
    loss = loss_func(out,y_train)
    
    if step%50 == 0:
        print("Epoch[{}/{}] loss:{}".format(step,config['epoch'],loss.numpy()))
    
    optimizer.zero_grad()
    
    loss.backward()
    
    optimizer.step()
    
    
    
```
